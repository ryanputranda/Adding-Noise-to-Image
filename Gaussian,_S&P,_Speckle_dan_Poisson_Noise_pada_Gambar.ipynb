```python
# Mengimpor library scikit image dan matplotlib untuk visualisasi gambarnya
import skimage
import matplotlib.pyplot as plt
```

```python
# Meload file gambar dan menampilkannya
img_path1="Ryan-Kristianto.jpg"
img1 = skimage.io.imread(img_path1)/255.0 # Menormalisasi gambar ke rentang RGB 0-255
plt.title('Original Asli')
plt.imshow(img1)
```

```python
# Memberi Gaussian Noise ke gambar dan menampilkannya
gimg = skimage.util.random_noise(img1, mode="gaussian")
plt.title('Gaussian Noise')
plt.imshow(gimg)
```

```python
# Memberi Poisson Noise ke gambar dan menampilkannya
gimg = skimage.util.random_noise(img1, mode="poisson")
plt.title('Poisson Noise')
plt.imshow(gimg)
```

```python
# Memberi S&P Noise ke gambar dan menampilkannya
gimg = skimage.util.random_noise(img1, mode="s&p")
plt.title('Salt & Pepper Noise')
plt.imshow(gimg)
```

```python
# Memberi Speckle Noise ke gambar dan menampilkannya
gimg = skimage.util.random_noise(img1, mode="speckle")
plt.title('Speckle Noise')
plt.imshow(gimg)
```
